```python
#Aquí van nuestros nombres
#Kelly Santos García
#Tania Monserratt Gopar Iniesta

#Lo que importamos para que el codigo funcione esta aquí
import re

#Empieza el codigo para programar la cosa esa para notación polaca inversa
#Definimos los operadores que usaremos, estos son los que vienen en la tarea
def is_operator(c):
    return c in ['+', '-', '*', '/', '**']

#Aquí definimos la prioridad de cada operador para evitar errores
def precedence(op):
    if op == '+' or op == '-':
        return 1
    if op == '*' or op == '/':
        return 2
    if op == '**':
        return 3
    return 0

#Aquí ponemos una clausula para que solo se ingresen caracteres válidos
def is_valid_expression(expression):
    # Verifica si la expresión contiene caracteres válidos
    valid_chars = set('0123456789+-*/() ** ')
    if not all(char in valid_chars for char in expression):
        return False
    # Verifica que haya espacios adecuados entre los elementos
    tokens = expression.split()
    # Expresión regular para detectar operadores de varios caracteres
    pattern = re.compile(r"^[0-9]+$|^[+\-*()/]+$")
    for token in tokens:
        if not pattern.match(token):
            return False
    # Verifica que no haya un número seguido de un paréntesis o un operador
    #sin espacio
    for i in range(len(expression) - 1):
        if (expression[i].isdigit() and expression[i + 1] in "()+-*/") or (expression[i] in "()+-*/" and expression[i + 1].isdigit()):
            if expression[i + 1] != ' ':
                return False

    return True

#Convierte la expresión expresión
def infix_to_postfix(expression):
    # Pila para almacenar los operadores
    stack = []
    # Cadena final para la expresión en notación polaca inversa
    output = []
    # Lista para almacenar los pasos
    steps = []

    # Separa la expresión en tokens
    tokens = expression.split()

    for token in tokens:
        if token.isnumeric():
            # Si es un número, agregarlo a la salida
            output.append(token)
            steps.append(f"Añadir {token} a la salida.")

        elif token == '(':
            # Si es un paréntesis izquierdo, apilarlo
            stack.append(token)
            steps.append(f"Apilar '{token}'.")

        elif token == ')':
            # Si es un paréntesis derecho, desapilar hasta encontrar el
            #paréntesis izquierdo
            while stack and stack[-1] != '(':
                output.append(stack.pop())
                steps.append(f"Desapilar y añadir a la salida.")
            stack.pop()  # Elimina el paréntesis izquierdo
            steps.append(f"Quitar '(' de la pila.")

        else:
            # Si es un operador
            while stack and precedence(stack[-1]) >= precedence(token):
                output.append(stack.pop())
                steps.append(f"Desapilar y añadir a la salida por mayor prioridad.")
            stack.append(token)
            steps.append(f"Apilar operador '{token}'.")

    # Desapila los demás operadores
    while stack:
        output.append(stack.pop())
        steps.append(f"Desapilar y añadir a la salida.")

    # Para guardar
    with open("Kelly y Monse programa.txt", "w") as file:
        for step in steps:
            file.write(step + "\n")

    return " ".join(output)

if __name__ == "__main__":
    while True:
        # Mensaje para pedir a la personita que introduzca una expresión
        expression = input("Holaaaa, introduce la expresión (o escribe 'bye' para cerrar el ciclo.) ")
        # Mensaje para salir del ciclo si el usuario ingresa 'bye'
        if expression.lower() == 'bye':
            print("Gracias por usar nuestro programa. ")
            break
        # Mensaje para validar que la entrada no esté vacía
        if not expression.strip():
            print("No ingresaste nada amigue :( vuelve a intentar plis. ")
            continue
        # Mensaje para validar que la expresión solo contenga caracteres
        #válidos y tenga espacios
        if not is_valid_expression(expression):
            print("O te falatan espacios o son caracteres invalidos, pero hay algo mal aquí amigue. ")
            continue

        # Convierte la expresión a notación polaca inversa
        result = infix_to_postfix(expression)

        # Mostrar el resultado
        print(f" Aquí la expresión en notación polaca inversa (si es que todo salio bien): {result}")
```
